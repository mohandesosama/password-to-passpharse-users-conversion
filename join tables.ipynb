```python
import pandas as pd
import numpy as np
import random
import matplotlib.pyplot as plt
%matplotlib inline
# we will get the same results every time we run the same program
random.seed(42)
df=pd.read_csv('Data/data_clean.csv')
df.head()
```

```python
df2=df

df2['intercept'] = 1
df2['ab_page'] = pd.get_dummies(df2['group'])['experiment']

countries_df = pd.read_csv('Data/countries.csv')
df_new = countries_df.set_index('student_id').join(df2.set_index('student_id'), how = 'inner')
df_new.head()
```

```python
df_new[['Health','IT','Business']] = pd.get_dummies(df_new['dept'])

df_new.head()
```

```python
import statsmodels.api as sm
log_mod2 = sm.Logit(df_new['satisfied'], df_new[['intercept','ab_page','Health','IT', 'Business']])
results2 = log_mod2.fit()
results2.summary()
```
